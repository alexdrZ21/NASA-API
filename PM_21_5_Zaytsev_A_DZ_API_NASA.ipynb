Марсоходы NASA каждый день ползают по Марсу, и каждый день делают снимки,
вывести три любых снимка любого марсохода при помощи открытого API NASA
за 2022 год, за дату, соответствующую Вашему дню рождения.

```python
import requests

api_key = "Gk7bC9f3hMI9EFLsMO4BuTDVZzdXh5csffaiYGgF"

url = f"https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?earth_date=2022-12-18&page=1&api_key={api_key}"

response = requests.get(url)

data = response.json()

photos = data["photos"][:3]

for i, photo in enumerate(photos):
    img_url = photo["img_src"]
    img_data = requests.get(img_url).content
    with open(f"mars_photo_{i}.jpg", "wb") as handler:
        handler.write(img_data)
        print(f"Снимок {i+1} сохранен.")
```
